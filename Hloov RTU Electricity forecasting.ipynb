```python
new_column = newdf[['Timestamp', 'RTU: Electricity']] 
new_column.dropna(inplace=True)
new_column.columns = ['ds', 'y'] 
new_column.tail()
```

```python
m = NeuralProphet(
    n_changepoints=100,
    trend_reg=2,
    yearly_seasonality=False,
    weekly_seasonality=False,
    daily_seasonality=False,
)
metrics = m.fit(new_column)
forecast = m.predict(new_column)
```

```python
future = m.make_future_dataframe(new_column, periods=1500)
forecast = m.predict(future)
forecast.tail()
```

```python
plot = m.plot(forecast)
```
